```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Creating the dataset
data = {"Month": np.arange(1, 13),"Production_BPD": [5000, 4800, 4600, 4450, 4300, 4150, 4000, 3850, 3700,3550, 3400, 3250],
        "Pressure_psi": [3500, 3450, 3400, 3350, 3300, 3250, 3200, 3150, 3100,3050, 3000, 2950] }
df = pd.DataFrame(data)
# Plot production trend
plt.scatter(df['Month'], df[ 'Pressure_psi'], color='red', label="Pressure")
plt.xlabel("Month")
plt.ylabel("Pressure")
plt.title("Pressure Trend")
plt.legend()
plt.show()
```

```python
# Prepare data for regression
X = df[["Month"]] # Independent variable
y = df["Production_BPD"] # Dependent variable
# Create and train the model
model = LinearRegression()
model.fit(X, y)
# Predict production for the next 6 months (Month 13-18)
future_months = np.array([[13], [14], [15], [16], [17], [18]])
predicted_production = model.predict(future_months)
# Display predictions
forecast = pd.DataFrame({"Month": [13, 14, 15, 16, 17, 18], 
"Predicted_Production_BPD": predicted_production})
print(forecast)
# Plot actual vs predicted production
plt.scatter(df["Month"], df["Production_BPD"], color='blue', 
label="Actual Production")
plt.plot(future_months, predicted_production, color='red', 
linestyle='dashed', label="Predicted Production")
plt.xlabel("Month")
plt.ylabel("Production_BPD")
plt.title("Oil Production Forecast")
plt.legend()
plt.show()
```

```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt

# Define the production control environment
class ProductionControlEnv:
    def __init__(self):
        self.state = None
        self.reset()
    
    def reset(self):
        self.state = random.randint(0, 20)  # Random initial state
        return self.state
    
    def step(self, action):
        # Actions: 0 = decrease, 1 = maintain, 2 = increase production
        if action == 0:
            self.state = max(0, self.state - 1)
        elif action == 2:
            self.state = min(20, self.state + 1)
        
        # Reward: Maximized at state 10
        reward = max(0, 100 - 10 * abs(self.state - 10))
        
        return self.state, reward

# Q-learning parameters
alpha = 0.1  # Learning rate
gamma = 0.9  # Discount factor
epsilon = 0.1  # Exploration rate
n_states = 21  # States 0 to 20
n_actions = 3  # Actions: decrease, maintain, increase

# Initialize Q-table
Q = np.zeros((n_states, n_actions))
env = ProductionControlEnv()
n_episodes = 1000  # Number of episodes
episode_length = 20  # Steps per episode
rewards_list = []  # Track rewards per episode
records = []  # Log each step

# Q-learning loop
for episode in range(n_episodes):
    state = env.reset()
    total_reward = 0

    for step in range(episode_length):
        # ε-greedy action selection
        if random.uniform(0, 1) < epsilon:
            action = random.choice([0, 1, 2])  # Explore
        else:
            action = np.argmax(Q[state, :])  # Exploit best action
        
        next_state, reward = env.step(action)
        total_reward += reward
        
        # Q-value update
        best_next_action = np.argmax(Q[next_state, :])
        Q[state, action] += alpha * (reward + gamma * Q[next_state, best_next_action] - Q[state, action])
        
        # Log step
        records.append([episode, step, state, action, reward, next_state])
        
        state = next_state

    rewards_list.append(total_reward)

# Convert logs to DataFrame
df_records = pd.DataFrame(records, columns=['Episode', 'Step', 'State', 'Action', 'Reward', 'Next State'])

# Plot learning curve
plt.plot(rewards_list)
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.title('Q-learning Performance')
plt.show()
```
